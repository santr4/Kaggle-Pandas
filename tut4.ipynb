Grouping and Sorting.

```python
import pandas as pd 
df = pd.read_csv('/home/aaranyak/Documents/10 year coding challenge/Kaggle-Pandas/Kaggle-Pandas/winemag-data-130k-v2.csv')
```

```python
df.head()
```

```python
df.groupby('points').points.count()
```

```python
df.groupby('points').points.min()
```

```python
df.groupby('winery').apply(lambda P : P.title.iloc[0])
```

```python
df.groupby(['country']).price.agg([len,min,max])
```

```python
countries_reviewed = df.groupby(['country','province']).description.agg([len])
countries_reviewed
```
